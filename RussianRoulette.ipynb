```python
import random
import time

bet = 0

class color:
   PURPLE = '\033[95m'
   CYAN = '\033[96m'
   DARKCYAN = '\033[36m'
   BLUE = '\033[94m'
   GREEN = '\033[92m'
   YELLOW = '\033[93m'
   RED = '\033[91m'
   BOLD = '\033[1m'
   UNDERLINE = '\033[4m'
   END = '\033[0m'
```

```python
def RussianRoulette():
    print(color.BOLD+ "Привет!" + color.END)
    time.sleep(1)
    
    print("""\n Welcome to a\n simulated version of \n
____ ____ ____ ____ ____ ____ ____ _________ ____ ____ ____ ____ ____ ____ ____ ____ 
||R |||u |||s |||s |||i |||a |||n |||       |||R |||o |||u |||l |||e |||t |||t |||e ||
||__|||__|||__|||__|||__|||__|||__|||_______|||__|||__|||__|||__|||__|||__|||__|||__||
|/__\|/__\|/__\|/__\|/__\|/__\|/__\|/_______\|/__\|/__\|/__\|/__\|/__\|/__\|/__\|/__\|

""")          
    time.sleep(3)
    
    print('First things first, you have to set your bet.')
    time.sleep(2)
    
    set_bet(bet)
```

```python
def set_bet(bet):
    print(f'You can bet any amount you wish starting with 1 and up to 1000.')
    possible_bets = range(1,1001)
    bet = int(input(f'Set your bet :'))

    if bet in possible_bets:
        time.sleep(1)
        print(f'\nYour bet is {bet}$.')
        
    if bet < 10:
        print(f'Only {bet}$ ? Come on...\n')
        time.sleep(2)
            
    elif bet >= 500:
        print('Gosh, we have a gambler here !\n')
        time.sleep(2)
        
    else:
        print(f'You need to choose a number between 1 & 1000.')

    
    
    print("Now, you have to select a weapon (this decision will affect the number of rounds you play).\n")
    time.sleep(3)
    choose_gun(bet)
```

```python
def choose_gun(bet):
    print('\n(a) = Classic Revolver (6 shots) \n(b) = S&W (7 shots, for the nervous ones) \n(c) = Beretta 93R (3 shots, for the braver ones)\n')
    
    is_wrong=True
    while is_wrong:
        gun = input("So, which gun do you want to use?")


        if gun == 'a':
            print("You selected the classic mode.")
            time.sleep(2)
            is_wrong=False
            ClassicRevolverGame(bet)

        elif gun == 'c':
            print("Okay... You're quite a gambler...")
            time.sleep(2)
            is_wrong=False
            Beretta93RGame(bet)

        elif gun == 'b':
            print("Playing it safe hein?")
            time.sleep(2)
            is_wrong=False
            SnWGame(bet)

        else:
            print("Please choose 'a', 'b' or 'c'.")


```

```python
def SnWGame(bet):
    
    shots = ['deadly', 'empty', 'empty', 'empty', 'empty', 'empty', 'empty']
    random.shuffle(shots)
    
    counter = 0
    new_bet = 0
    
    while len(shots) != 0:
        
        bam = shots.pop()
        counter += 1
        
        print("\nRound #", counter)
        print("Your shot was", bam)
        
        if bam == 'deadly':
            print("You died.")
            time.sleep(1)
            print("And your bet too.")
            time.sleep(1)
            print("GAME OVER")
            break
            
        if bam == 'empty':
            
            new_bet += bet*2
            time.sleep(1)
            
            print(f'Nice! Your bet has doubled and you now have {new_bet}.')
            
            time.sleep(1)
            
            decision = input('Do you want to shoot again and try to increase your bet once more? y/n\n')
            
            if decision == "y":
                bam
                
            if decision == "n":
                print(f'You choose to stop here. You won {new_bet}$ and you are still alive, congratulations!')
                break
```
